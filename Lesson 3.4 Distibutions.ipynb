```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rand_dist = np.random.uniform(5,10,1000) 
mean = rand_dist.mean()
std = rand_dist.std()
plt.hist(rand_dist, color='c')
plt.axvline(mean, color='b', linewidth=2, linestyle='solid')
plt.axvline(mean + std, color='b', linewidth=2, linestyle='dotted')
plt.axvline(mean - std, color='b', linewidth=2, linestyle='dotted')
```

## The Uniform distribution appears to be spread evenly across the distribution.

```python
rand_dist = np.random.exponential(5,10000) 
mean = rand_dist.mean()
std = rand_dist.std()
plt.hist(rand_dist, color='c', bins="auto")
plt.axvline(mean, color='b', linewidth=2, linestyle='solid')
plt.axvline(mean + std, color='b', linewidth=2, linestyle='dotted')
plt.axvline(mean - std, color='b', linewidth=2, linestyle='dotted')
```

## The Exponential distribution appears to become exponentially larger toward the front end of the distribution.

```python
rand_dist = np.random.geometric(.20,1000) 
mean = rand_dist.mean()
std = rand_dist.std()
plt.hist(rand_dist, color='c', bins="auto")
plt.axvline(mean, color='b', linewidth=2, linestyle='solid')
plt.axvline(mean + std, color='b', linewidth=2, linestyle='dotted')
plt.axvline(mean - std, color='b', linewidth=2, linestyle='dotted')
```

## The Geometric distribution defines the number of tries it takes to achieve a given result, given the probably p.  In this case prob .20 takes 5 tries on average.

```python
rand_dist = np.random.chisquare(5,1000) 
mean = rand_dist.mean()
std = rand_dist.std()
plt.hist(rand_dist, color='c', bins="auto")
plt.axvline(mean, color='b', linewidth=2, linestyle='solid')
plt.axvline(mean + std, color='b', linewidth=2, linestyle='dotted')
plt.axvline(mean - std, color='b', linewidth=2, linestyle='dotted')
```

## The chi-square is a normal distribution slightly skewed left.
## It is the result of squaring random (p) normal distributions with mean 0 var 1 and summing the result.   In this case p was 5.

```python
rand_dist = np.random.binomial(10,.1,1000) 
mean = rand_dist.mean()
std = rand_dist.std()
plt.hist(rand_dist, color='c', bins="auto")
plt.axvline(mean, color='b', linewidth=2, linestyle='solid')
plt.axvline(mean + std, color='b', linewidth=2, linestyle='dotted')
plt.axvline(mean - std, color='b', linewidth=2, linestyle='dotted')
```

## The binomial represents the number of successes given n tries with a probability of success p.
##  In this case n was 10, and p .10.
##  So as expected, I'm most likely to get 1 success out of 10 tries. However, in some cases I got 0, 2, 3, and 4 successes.

```python
rand_dist = np.random.triangular(0,5,20,100000) 
mean = rand_dist.mean()
std = rand_dist.std()
plt.hist(rand_dist, color='c', bins="auto")
plt.axvline(mean, color='b', linewidth=2, linestyle='solid')
plt.axvline(mean + std, color='b', linewidth=2, linestyle='dotted')
plt.axvline(mean - std, color='b', linewidth=2, linestyle='dotted')
```

## The triangular represents a distribution that peaks at the mode, with a minimum and a maximum.
##  In this case the mode 5, min 0, max 20

```python
norm1 = np.random.normal(5,.5,10000)
norm2 = np.random.normal(10,1,10000)
norm3 = norm1 + norm2
mean = norm3.mean()
std = norm3.std()
plt.hist(norm3, color='c', bins="auto")
plt.axvline(mean, color='b', linewidth=2, linestyle='solid')
plt.axvline(mean + std, color='b', linewidth=2, linestyle='dotted')
plt.axvline(mean - std, color='b', linewidth=2, linestyle='dotted')
```

## When 2 normal distribitions are added together, the resulting distribution has a mean and std reflecting the sum of the 2.
